```python
# Given the list [10, 15, 20, 25, 30], here is a function that returns the square of only even numbers

def square_even(numbers):

  even_squares = []
  for number in numbers:
    if number % 2 == 0:
      even_squares.append(number**2)
  return even_squares

# Example usage with the provided list:
my_numbers = [10, 15, 20, 25, 30]
result = square_even(my_numbers)
print(result)
# expected  Output- [100, 400, 900]
```

```python
import pandas as pd

data = {
    "Region": ["North", "South", "North", "East", "South", "East"],
    "Product": ["A", "B", "A", "B", "A", "B"],
    "Sales": [100, 150, 200, 50, 75, 125],
    "Price": [10, 15, 20, 5, 8, 12]
}

df = pd.DataFrame(data)
print(df)
```

general/example  dataframe for showing how the following  tasks perform.

 1. The .groupby() function splits a DataFrame into groups based on one or more columns, applies a function (e.g., sum, mean, count) to each group, and combines the results into a new DataFrame.

```python
grouped = df.groupby("Region").agg({
    # Total sales per region
    "Sales": "sum",

    # AVG. price per region
    "Price": "mean"
}).reset_index()

print(grouped)
# Multi-Column Grouping:
df.groupby(["Region", "Product"]).sum()
```

```python
# first task: Create a list or column that has a low and a high std. dev, and compare.

import pandas as pd
import numpy as np

# Create a DataFrame with a low standard deviation
low_std_data = {'values': np.random.normal(loc=50, scale=5, size=100)}
low_std_df = pd.DataFrame(low_std_data)

# Create a DataFrame with a high standard deviation
high_std_data = {'values': np.random.normal(loc=50, scale=20, size=100)}
high_std_df = pd.DataFrame(high_std_data)


print("Low Standard Deviation:")
print(low_std_df.describe())
print("\nHigh Standard Deviation:")
print(high_std_df.describe())


```

 1: Standard deviation is the  measurement of  how spread out values are in a dataset are.
 - A high standard deviation means the data points are more spread out from the mean (average)
 - A low standard deviation means the data points are clustered closely around the mean.
